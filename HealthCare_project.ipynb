```python
# Convert date columns to datetime
HC['Date of Admission'] = pd.to_datetime(HC['Date of Admission'])
HC['Discharge Date'] = pd.to_datetime(HC['Discharge Date'], errors='coerce')
HC['Date of Admission']
HC['Discharge Date'])
```

```python
# Group by medical condition and calculate average length of stay
recovery_trends = HC.groupby('Medical Condition')['Length_of_Stay'].mean().reset_index()
recovery_trends 
```

```python
recovery_trends = recovery_trends.sort_values(by='Length_of_Stay')
recovery_trends
```

```python
# Plot patient recovery trends
plt.figure(figsize=(10,6))
plt.bar(recovery_trends['Medical Condition'], recovery_trends['Length_of_Stay'])
plt.xlabel('Medical Condition')
plt.ylabel('Average Length of Stay')
plt.title('Patient Recovery Trends')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print(recovery_trends)
```

```python
# Create age groups
HC['Age_Group'] = pd.cut(HC['Age'], bins=[0, 20, 60, np.inf], labels=['Below 20', '21-60', 'Above 60'], include_lowest=True, right=False)
HC['Age_Group']
```

```python
# Group by age group and medical condition, calculate average length of stay
recovery_trends_below_20 = HC[HC['Age_Group'] == 'Below 20'].groupby('Medical Condition')['Length_of_Stay'].mean().reset_index().sort_values(by='Length_of_Stay', ascending=False).head(10)
recovery_trends_21_60 = HC[HC['Age_Group'] == '21-60'].groupby('Medical Condition')['Length_of_Stay'].mean().reset_index().sort_values(by='Length_of_Stay', ascending=False).head(10)
recovery_trends_above_60 = HC[HC['Age_Group'] == 'Above 60'].groupby('Medical Condition')['Length_of_Stay'].mean().reset_index().sort_values(by='Length_of_Stay', ascending=False).head(10)

recovery_trends_below_20

recovery_trends_21_60

recovery_trends_above_60
```

```python
# Plot patient recovery trends
fig, axs = plt.subplots(3, figsize=(10, 15))
axs[0].bar(recovery_trends_below_20['Medical Condition'], recovery_trends_below_20['Length_of_Stay'])
axs[0].set_title('Patient Recovery Trends by Medical Condition (Below 20)')
axs[0].set_xlabel('Medical Condition')
axs[0].set_ylabel('Average Length of Stay')
axs[0].tick_params(axis='x', rotation=90)

axs[1].bar(recovery_trends_21_60['Medical Condition'], recovery_trends_21_60['Length_of_Stay'])
axs[1].set_title('Patient Recovery Trends by Medical Condition (21-60)')
axs[1].set_xlabel('Medical Condition')
axs[1].set_ylabel('Average Length of Stay')
axs[1].tick_params(axis='x', rotation=90)

axs[2].bar(recovery_trends_above_60['Medical Condition'], recovery_trends_above_60['Length_of_Stay'])
axs[2].set_title('Patient Recovery Trends by Medical Condition (Above 60)')
axs[2].set_xlabel('Medical Condition')
axs[2].set_ylabel('Average Length of Stay')
axs[2].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()

print(recovery_trends_below_20)
print(recovery_trends_21_60)
print(recovery_trends_above_60)
```

```python
# Group by doctor and calculate average length of stay
doctor_performance = HC.groupby('Doctor')['Length_of_Stay'].mean().reset_index().sort_values(by='Length_of_Stay')
doctor_performance
```

```python
# Rank the doctors based on their average length of stay
doctor_performance = doctor_performance.sort_values(by='Length_of_Stay')
doctor_performance 
```

```python
# Get the top 10 performing doctors
top_doctors = doctor_performance.nsmallest(10, 'Length_of_Stay')
top_doctors
```

```python
# Get the bottom 10 performing doctors
bottom_doctors = doctor_performance.nlargest(10, 'Length_of_Stay')
bottom_doctors 
```

```python
# Plot the top-performing doctors
plt.figure(figsize=(10,6))
plt.bar(top_doctors['Doctor'], top_doctors['Length_of_Stay'])
plt.xlabel('Doctor')
plt.ylabel('Average Length of Stay')
plt.title('Top-Performing Doctors')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


print(top_doctors)
```

```python
# Plot the bottom-performing doctors
plt.figure(figsize=(10,6))
plt.bar(bottom_doctors['Doctor'], bottom_doctors['Length_of_Stay'])
plt.xlabel('Doctor')
plt.ylabel('Average Length of Stay')
plt.title('Bottom-Performing Doctors')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print(bottom_doctors)
```
